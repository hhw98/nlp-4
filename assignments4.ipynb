```python
import nltk
import re
import jieba
import gensim
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#STOP_WORDS = nltk.corpus.stopwords.words()

def build_corpus(val):
    corpus = []
    "remove chars that are not letters or numbers, downcase, then remove stop words"
    #regex = re.compile('[\u4e00-\u9fff]+')
    sentence = re.findall('[\u4e00-\u9fff]+',val) 
            
    sentence = "".join(sentence)
    sentence = HanziConv.toSimplified(sentence)
    #corpus.append(jieba.lcut(sentence))
    return ' '.join(jieba.cut(sentence))

```

```python
cutted_file = open('cutted_wiki.txt','w',encoding = 'utf-8')
```

```python
for dirs in os.listdir("wikiextractor-master/text/"):
    for file in os.listdir("wikiextractor-master/text/"+dirs):
        print(os.path.join("wikiextractor-master/text/", dirs, file))
        for line in open(os.path.join("wikiextractor-master/text/", dirs, file),encoding = 'utf-8'):
            cutted_file.write(build_corpus(line)+'\n')
```

```python
import gensim
text = gensim.models.word2vec.LineSentence('cutted_wiki.txt')
model = gensim.models.Word2Vec(text, size=35, workers=8)
```

```python
model.save("model/word2vec.model")
```

```python
def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(800, 800)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(50, 20),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
tsne_plot(model)
```
